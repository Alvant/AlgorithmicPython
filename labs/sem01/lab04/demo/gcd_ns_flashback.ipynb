```python
# 10 -> ? (<= 10) (например, в 7-ю)

base = 7

assert base <= 10

num = 7 ** 3  # 14 + 1 = 2 * 7 + 1
res = ""

while num > 0:
    d = num % base      # цифра в base-представлении числа num
    
    # ... тут ещё действие для 16-ой и т.п. ("красивая цифра")
    
    res = str(d) + res  # наращиваем число справа налево (единицы, десятки и т.д.)
    num = num // base   # "всё кроме цифр" и сдвинутое вправо

if len(res) == 0:
    res = "0"

print(res)
```

## Полный перебор

(на примере задачи поиска НОД двух чисел)

```python
"""
Сдать решение задачи A-Алгоритм Евклида
Полный балл:	50
Штраф за посылку:	1
Ограничение времени:	1 с
Ограничение реального времени:	1 с
Ограничение памяти:	64M

Необходимо найти НОД двух чисел, используя алгоритм Евклида.
Формат входных данных

На вход подаются два натуральных числа, по числу в новой строке.
Формат выходных данных

Одно число - НОД входных чисел. 

"""
```

```python
a = int(input())
b = int(input())
```

```python
print(a, b)
```

```python
gcd = 1

for d in range(2, min(a, b) + 1):
    if a % d == 0 and b % d == 0:
        gcd = d

print(gcd)
```

То же самое, но красивенько в виде функции:

```python
def gcd(a: int, b: int) -> int:
    res = 1

    for d in range(2, min(a, b) + 1):
        if a % d == 0 and b % d == 0:
            res = d
    
    return res
```

```python
gcd(a, b)
```

### Дополнение 1: "Канонический" (рекурсивный) вариант решения

```python
def gcd_rec(a: int, b: int) -> int:
    # a - max
    # b - min
    # a = c * b + r
    # (a, b) -> (b, r) -> ... -> r = 0
    
    # a, b = max(a, b), min(a, b)
    
    if b == 0:
        return a
    
    return gcd_rec(b, a % b)
```

```python
gcd_rec(a, b)
```

## Системы счисления: Флешбэк

```python
"""Сдать решение задачи E-Считывание числа в системе счисления майя
Полный балл:	150
Штраф за посылку:	1
Ограничение времени:	1 с
Ограничение реального времени:	1 с
Ограничение памяти:	64M

Система счисления майя основывалась на 20-ричной системе счисления.

Разряд формируется следующим образом. Для записи нуля использовалась ракушка "@". Для конструирования числа разряда использовались точки ".", и палочки "|". Точка добавляла к числу разряда 1, а палочка 5. Например, число 13 помещается в один разряд и выглядит следующим образом: "...||" = 3 + 2*5.

Для отделения разрядов мы воспользуемся пробелом.

Вам предстоит считать число майя и перевести его в десятичную систему счисления.
Формат входных данных

Одна строка - число, записанное в системе счисления майя. 

"""
```

```python
s = input()

l = s.split()
```

```python
l
```

```python
def int_from_maya(s):
    point = 1
    stick = 5
    
    return s.count(".") * point + s.count("|") * stick
```

```python
int_from_maya("....")
```

```python
int_from_maya(".|")
```

```python
# 1. Цифры -> в нормальный вид

l = [
    int_from_maya(e) for e in l
]
```

```python
l
```

```python
l[::-1]
```

```python
list(enumerate(["a", "b", "c"]))
```

```python
# 2. Перевод из "какой-то" в десятичную

base = 20
res = 0

for i, d in enumerate(l[::-1]):
    res += d * base ** i
```

### Дополнение: Как G (в другую сторону)

```python
n = int(input())
```

```python
print(n)
```

```python
# 1. Перевод из десятичной в "какую-то"

base = 20
l = []

while n > 0:
    d = n % base  # Текущий разряд
    l.append(d)
    
    n = n // base  # "Двигаем" разряды вправо
```

```python
l
```

```python
def maya_from_int(n):
    if n == 0:
        return "@"

    point = 1
    stick = 5
    
    stick_count = n // stick
    point_count = n % stick
    
    return "." * point_count + "|" * stick_count
```

```python
# 2. "Причёсываем" разряды

l = [
    maya_from_int(e) for e in l
]
```

```python
l
```

```python
print(" ".join(l[::-1]))
```
